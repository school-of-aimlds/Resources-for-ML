# Operators in Python
- Arithmetic Operators                     
- Assignment Operators               
- Comparison Operators                 
- Logical Operators               
- Identity Operators                 
- Membership Operaors              
- Bitwise Operators

```python
#Assignment Operator

num_1=10
num_2=5

#Addition
sum = num_1 + num_2
print('Sum =',sum)

#Subtraction
diff = num_2 - num_1
print('Difference =',diff)

#Multiplication
prod = num_1 * num_2
print('Product =',prod)

#Division
div = num_2 / num_1
print('Division =',div)

#Floor
floor = num_1 // num_2
print('Floor =',floor)            # retruns integer quotient 

#Exponent
exp = num_1**num_2
print('Exponent =',exp)           # returns the power

#Modulus
mod = num_1 % num_2
print('Modulus =',mod)            # returns remainder
```

```python
# Assignment Opeartors

a = 5                  # "=" is assignment operator
print(a)

a = 5
a += 5                 # "+=" means a= a+5
print(a)

b = 10
b -= 2                 # "-=" means b= b-2
print(b)

c = 3
c *= 4                # c= c*4
print(c)

c = 3
c **= 4               # c= c^4
print(c)

c = 5
c /= 2                # c= c/2
print(c)

c = 12
c %=3                # c= c%3
print(c)
```

```python
# Comparison Operators - returns either true or false

a = 5
b = 10

print(a == b)       #equal to
print(a != b)       #not equal to
print(a > b)        #greater than
print(a < b)        #less than
print(a <= b)       #less than equal to
print(a >= b)       #greater than equal to
```

```python
# Logical Operators

a = 10

#and
print( a>20 and a<10)
print( a <=10 and a>8)       #True if both are true

#or
print( a>9 or a>30)
print( a>20 or a==12)        #True if either one is true

#not
print( not(a>5))             #returns opposite
```

```python
# Identity Opeartors

x = 5
y = 5

#is
print(x is y)         #Returns true if both are same

#is not
print(x is not y)
```

```python
# Membership Operators

a = 5
b = 10
c = [1,2,3,4,5]

# in
print(a in c)
print(a not in c)            #true if element is in the object
 
# not in
print(b in c)
print(b not in c)
```

```python
#Bitwise Operators - Check each bit at each place

a=3       #3=011
b=6       #6=110

# and- returns 1 if both are 1
print(a&b)           #010=2

# or- returns 1 if either is 1
print(a|b)           #111=7

# X-OR- returns 1 if only one of the two bits is 1
print(a^b)           #101=5

# Not- ~n=-(n+1)
print(~a)
print(~b)
```
