Some of the major Pros of Matplotlib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

## Plotting using matplotlib 
### 1. Scatterplot
A scatter plot is a type of plot that displays individual data points on a two-dimensional plane. It is useful for visualizing the relationship between two variables. Each data point is represented as a dot, and the position of the dot on the plot corresponds to the values of the two variables. Scatter plots are commonly used to identify patterns or correlations between variables, to detect outliers, or to visualize the distribution of data.

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')   # xlabel- sets the label of x-axis
plt.ylabel('Y axis')   # ylabel- sets the label of y-axis
plt.title('Graph in 2D')     # title- sets the title of the plot
plt.savefig('Test.png')   # savefig- will save the graph plot
```

#### plt.scatter(x, y, c='g'): 
This line creates a scatter plot with the given data points x and y. Each point in the scatter plot corresponds to a pair of values from x and y. The parameter c='g' specifies the color of the points, in this case, 'g' stands for green. You can use any valid color name or a hexadecimal code to specify the color.
since, here x is values from 0 to 10 and y is sin(x), so it makes a graph of that

### 2. plt.plot() function
The plt.plot() function in matplotlib is used to create a line plot. It takes in one or more pairs of arrays or lists representing the x and y coordinates of data points. Additionally, it allows you to specify various parameters to customize the appearance of the plot, such as line style, color, marker style, line width, and marker size.

```python
## plt plot
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

### 3. Subplots
plt.subplot() is a function in the Matplotlib library of Python used to create subplots within a single figure. Subplots are multiple plots in a single figure, arranged in a grid pattern. This allows you to compare different datasets or views of data side by side, making it easier to visualize relationships or patterns.

The plt.subplot() function is used to specify the layout of the subplots within the figure. It takes three integer arguments:

#### plt.subplot(num_rows, num_cols, plot_num): 
This function specifies the layout of subplots in a grid with num_rows rows and num_cols columns, and it creates a subplot at position plot_num.
- num_rows: The number of rows in the subplot grid.
- num_cols: The number of columns in the subplot grid.
- plot_num: The position of the subplot in the grid. Subplots are numbered starting from 1, arranged row by row from left to right.

```python
## Creating Subplots

plt.subplot(2,2,1)     
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(z,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

Each plt.subplot() call defines the position of the subplot within the grid. The first argument is the number of rows, the second argument is the number of columns, and the third argument is the index of the current subplot. The subplots are numbered starting from 1, and they are arranged row by row from left to right. In this case, the subplots are arranged in a 2x2 grid, so there are four subplots, numbered from 1 to 4.

```python
# Sample data for plotting
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Creating a 2x3 grid of subplots
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title('Plot 1')

plt.subplot(2, 3, 2)
plt.plot(x, [i**2 for i in x], 'r--')
plt.title('Plot 2')

plt.subplot(2, 3, 3)
plt.plot(x, [i**3 for i in x], 'g-.')
plt.title('Plot 3')

plt.subplot(2, 3, 4)
plt.plot(x, [i**0.5 for i in x], 'ko')
plt.title('Plot 4')

plt.subplot(2, 3, 5)
plt.plot(x, [i**1.5 for i in x], 'b')
plt.title('Plot 5')

plt.subplot(2, 3, 6)
plt.plot(x, [i**2.5 for i in x], 'm')
plt.title('Plot 6')

plt.tight_layout()  # Adjust subplot layout to make it neater. The tight_layout() function adjusts the spacing between subplots to prevent overlapping elements and improve the layout of the plot.
plt.show()
```

### 4. Bar plot
A bar plot (or bar chart) is a type of plot that presents categorical data with rectangular bars. The lengths of the bars represent the values of the data points they represent. Bar plots are useful for comparing data across categories or groups. In this example, two sets of bars are plotted side by side, each representing different datasets. The first set is plotted using the default color, while the second set is explicitly set to green.

```python
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y)   # by default color is blue
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

### 5. Histogram
A histogram is a graphical representation of the distribution of numerical data. It consists of a series of adjacent rectangles (bars), where each bar represents the frequency (or count) of data points within a specific range or "bin" of values. Histograms are useful for visualizing the underlying distribution of a dataset, including its central tendency, variability, and shape.

```python
b = np.array([10,11,10,12,20,30,25,15]) 
plt.hist(b) 
plt.title("histogram") 
plt.show()
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

### 6. box-plot

A boxplot, also known as a box-and-whisker plot, provides a visual summary of the distribution of a dataset. It displays several descriptive statistics such as the median, quartiles, and potential outliers.

Here's what a boxplot typically tells us:
1. Median (Q2/50th Percentile): The middle value of the dataset. It represents the point where half of the data is below and half is above.
2. Quartiles (Q1, Q3): The dataset is divided into four equal parts. Q1 (lower quartile) is the value below which 25% of the data falls, and Q3 (upper quartile) is the value below which 75% of the data falls.
3. Interquartile Range (IQR): The range between the first and third quartiles (Q3 - Q1). It describes the middle 50% of the data.
4. Whiskers: Lines extending from the box that represent the range of the data, excluding outliers. The whiskers typically extend to 1.5 times the IQR from the quartiles.
5. Outliers: Data points that fall outside the whiskers. These points are typically plotted individually.
6. Box length: The length of the box represents the spread of the middle 50% of the data. A longer box indicates a larger spread.

By observing a boxplot, you can quickly grasp key aspects of the distribution of your data, such as its central tendency, spread, and presence of outliers. It's particularly useful for comparing distributions between different groups or datasets.

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);
```

```python
data = [10,15,20,22,12,36]

# Plot a boxplot
plt.boxplot(data)
plt.title('Boxplot')
plt.ylabel('Value')
plt.show()
```

1. Median (Q2): The median value of the data is approximately 17.5. This divides the dataset into two halves, with half of the values falling below and half above this point.

2. Quartiles (Q1, Q3):
- Q1 (lower quartile) is approximately 12.25. This indicates that 25% of the data falls below this value.
- Q3 (upper quartile) is approximately 21.5. This means that 75% of the data falls below this value.
3. Interquartile Range (IQR): The IQR is approximately 9.25 (Q3 - Q1). It represents the middle 50% of the data.
4. Whiskers: The whiskers extend from Q1 to Q3. They indicate the range of the data, excluding outliers.
- The upper whisker extends to the highest value within 1.5 times the IQR above Q3, which in this case is approximately 21.5+1.59.25=35.375
- Since there's a value of 36 in the dataset, it's considered an outlier and is plotted individually.
- he lower whisker extends down to the lowest value within 1.5 times the IQR below Q1.
5. Box length: The length of the box represents the spread of the middle 50% of the data. A longer box indicates a larger spread.

### 7. Pie-Chart
A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. Each slice represents a proportionate part of the whole, and the size of each slice is proportional to the quantity it represents. Pie charts are useful for visualizing the relative sizes of different categories or components within a dataset.

In a pie chart:
- The entire circle represents the total data set or 100%.
- Each slice represents a subset of the data, and the size of the slice corresponds to the proportion of the total data that it represents.
- Typically, the slices are labeled with the category or component they represent, and sometimes the numerical value or percentage is also displayed.

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (1, 0.5, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
